```python
import numpy as np
from tensorflow import keras 
from keras.datasets import mnist
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print("training images shape: ", train_images.shape)
print("training labels shape: ", train_labels.shape)
print("test images shape: ", test_images.shape)
print("test labels shape: ", test_labels.shape)
```

```python
train_images = train_images.reshape(-1, 784)
test_images = test_images.reshape(-1, 784)
```

```python
print("training images shape: ", train_images.shape)
print("training labels shape: ", train_labels.shape)
print("test images shape: ", test_images.shape)
print("test labels shape: ", test_labels.shape)
```

```python
train_images[0]
```

```python
test_images[0]
```

```python
#normalizing data
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
train_images[0]
```

```python
model = Sequential()
model.add(Dense(64, activation = 'relu', input_shape = (784,)))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_images, to_categorical(train_labels), epochs = 10, batch_size = 32)


#observation1: if we dont include batch_size, then for 10 epochs we got train accuracy 93
#observation2: if we include batch_size = 32, then for 10 epochs we got train accuracy 96
#observation3: conclusion is that by including batch size, train accuracy got increased
```

```python
#evaluate the model

model.evaluate(test_images, to_categorical(test_labels))
```

```python
#save the model to disk

model.save_weights('ffnn_mnist.h5')
```
